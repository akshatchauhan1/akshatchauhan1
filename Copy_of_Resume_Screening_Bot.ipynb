```python
from sentence_transformers import SentenceTransformer, util
import os
import torch
```

```python
model = SentenceTransformer('all-MiniLM-L6-V2');
model
```

```python
job_description = """
We are looking for a Python Developer with experience in machine learning,
NLP,deep learning, reinforcement learning and REST API development. The ideal candidate has worked with TensorFlow,
Flask, and has a strong understanding of data structures and algorithms.
"""
```

```python
resume_folder = "resumes/"  # Create this folder and add resumes as .txt files
```

```python
resume_scores = {}
```

```python
job_embedding = model.encode(job_description, convert_to_tensor=True)
```

```python
import os

# Create the 'resumes' folder if it doesn't exist
if not os.path.exists("resumes"):
    os.makedirs("resumes")
```

```python
# Read and compare each resume
for resume_file in os.listdir(resume_folder):
    if resume_file.endswith(".txt"):
        with open(os.path.join(resume_folder, resume_file), "r", encoding="utf-8") as f:
            resume_text = f.read()

        resume_embedding = model.encode(resume_text, convert_to_tensor=True)
        similarity = util.pytorch_cos_sim(job_embedding, resume_embedding).item()
        resume_scores[resume_file] = similarity

```

```python
sorted_resumes = sorted(resume_scores.items(), key=lambda x: x[1], reverse = True)
```

```python
print("\nTop Matching Resumes:")
for resume, score in sorted_resumes[:5]:
    print(f"{resume : scores :.(4f)}")
```
